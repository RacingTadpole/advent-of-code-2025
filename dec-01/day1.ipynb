```python
with open('input.txt', 'r') as f:
    lines = f.readlines()

lines = [line.rstrip('\n') for line in lines]
```

```python
rotations = [int(line[1:]) * (-1 if line[0] == 'R' else 1) for line in lines]
```

```python
def count_zeros(start_position, rotations):
    """
    >>> count_zeros(50, [68, 30, -48, 5, -60, 55, 1, 99, -14, 82])
    3
    """
    def rotate(pos, rot):
        return (pos + rot) % 100

    count = 0
    position = start_position
    for rotation in rotations:
        position = rotate(position, rotation)
        if position == 0:
            count += 1
    
    return count
```

```python
print(count_zeros(50, rotations))
```
