```python
import doctest
```

```python
with open('input.txt', 'r') as f:
    lines = f.readlines()

lines = [line.rstrip('\n') for line in lines]
```

```python
rotations = [int(line[1:]) * (-1 if line[0] == 'L' else 1) for line in lines]
```

# Part 1

```python
def rotate(pos, rot):
    """
    >>> rotate(50, 68)
    18
    >>> rotate(18, 30)
    48
    >>> rotate(48, -48)
    0
    >>> rotate(20, -505)
    15
    """
    return (pos + rot) % 100

def count_zeros_after_rotations(start_position, rotations):
    """
    >>> count_zeros_after_rotations(50, [-68, -30, 48, -5, 60, -55, -1, -99, 14, -82])
    3
    """
    count = 0
    position = start_position
    for rotation in rotations:
        position = rotate(position, rotation)
        if position == 0:
            count += 1
    
    return count

doctest.testmod()
```

```python
print(count_zeros_after_rotations(50, rotations))
```

# Part 2

```python
def count_zeros_during_rotation(pos, rot):
    """
    >>> count_zeros_during_rotation(50, 68)
    (18, 1)
    >>> count_zeros_during_rotation(20, -25)
    (95, 1)
    >>> count_zeros_during_rotation(20, -125)
    (95, 2)
    >>> count_zeros_during_rotation(10, 90)
    (0, 1)
    >>> count_zeros_during_rotation(10, -10)
    (0, 1)
    >>> count_zeros_during_rotation(0, -10)
    (90, 0)
    >>> count_zeros_during_rotation(0, -101)
    (99, 1)
    """
    target = pos + rot
    if target > 0:
        zeros_crossed = target // 100
    elif pos == 0:
        zeros_crossed = -target // 100
    else:
        zeros_crossed = -target // 100 + 1
    
    return target % 100, zeros_crossed

def count_zeros_during_rotations(start_position, rotations):
    """
    >>> count_zeros_during_rotations(50, [-68, -30, 48, -5, 60, -55, -1, -99, 14, -82])
    6
    """
    count = 0
    position = start_position
    for rotation in rotations:
        position, this_count = count_zeros_during_rotation(position, rotation)
        count += this_count

    return count

doctest.testmod()
```

```python
count_zeros_during_rotations(50, rotations)
```
