```python
import doctest
```

```python
with open('input.txt', 'r') as f:
    lines = f.readlines()

banks = [line.replace('\n', '') for line in lines]
```

# Part 1

```python
def get_joltage(digits: str) -> int:
    """
    >>> get_joltage('987654321111111')
    98
    >>> get_joltage('811111111111119')
    89
    >>> get_joltage('234234234234278')
    78
    >>> get_joltage('818181911112111')
    92
    """
    # Find the max number before the last number
    first_digit = max(digits[:-1])
    first_digit_position = digits.index(first_digit)
    second_digit = max(digits[first_digit_position + 1:])

    return int(first_digit) * 10 + int(second_digit)

doctest.testmod()
```

```python
def get_total_joltage(joltages: list(str), get_joltage = get_joltage) -> int:
    """
    >>> get_total_joltage(['987654321111111', '811111111111119','234234234234278', '818181911112111'])
    357
    """
    
    total = 0
    for digits in joltages:
        total += get_joltage(digits)
    return total

doctest.testmod()
```

```python
get_total_joltage(banks, get_joltage)
```

# Part 2

```python
def get_big_joltage(digits: str, subsize = 12) -> int:
    """
    >>> get_big_joltage('987654321111111')
    987654321111
    >>> get_big_joltage('811111111111119')
    811111111119
    >>> get_big_joltage('234234234234278')
    434234234278
    >>> get_big_joltage('818181911112111')
    888911112111
    """
    start_index = 0
    result: str = ''
    for digit_index in range(subsize):
        candidate_digits = digits[start_index:(digit_index - subsize + 1)] if digit_index - subsize + 1 != 0 else digits[start_index:]
        max_digit = max(candidate_digits)
        result += max_digit
        start_index = digits[start_index:].index(max_digit) + start_index + 1

    return int(result)

doctest.testmod()
```

```python
get_total_joltage(banks, get_big_joltage)
```
