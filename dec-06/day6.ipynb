```python
import math, doctest
```

```python
with open('input.txt', 'r') as f:
    raw_lines = f.readlines()

lines = [line.replace('\n', '').split() for line in raw_lines]
```

```python
operations = lines[-1]
numbers = [[int(n) for n in line] for line in lines[:-1]]
# Check the lengths all make sense
len(operations), [len(line) for line in numbers]
```

# Part 1

```python
def squid_maths(operations, numbers):
    """
    >>> squid_maths(['*','+','*','+'], [[123,328,51,64], [45,64,387,23], [6,98,215,314]])
    4277556
    """
    total = 0

    for index, operation in enumerate(operations):
        if operation == '+':
            total += sum(line[index] for line in numbers)
        elif operation == '*':
            total += math.prod(line[index] for line in numbers)
        else:
            print(f'Invalid operation {operation} on line {index}')
    
    return total

doctest.testmod()
```

```python
squid_maths(operations, numbers)
```
