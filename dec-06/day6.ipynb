# Part 1

```python
def squid_maths(operations, numbers):
    """
    >>> squid_maths(['*','+','*','+'], [[123,328,51,64], [45,64,387,23], [6,98,215,314]])
    4277556
    """
    total = 0

    for index, operation in enumerate(operations):
        if operation == '+':
            total += sum(line[index] for line in numbers)
        elif operation == '*':
            total += math.prod(line[index] for line in numbers)
        else:
            print(f'Invalid operation {operation} on line {index}')
    
    return total

doctest.testmod()
```

```python
squid_maths(operations, numbers)
```

```python
with open('input.txt', 'r') as f:
    raw_lines = f.readlines()

lines = [line.replace('\n', '') for line in raw_lines]

number_lines = lines[:-1]
operations = lines[-1]

[len(line) for line in lines] # Note these are all padded so they are the same
```

```python
def new_maths(operations, number_lines):
    current_operation = ''
    totals = []
    this_total = 0
    for column_index in range(len(operations)):
        if operations[column_index] in ['+', '*'] or column_index == len(operations):
            # Save the last result and start afresh
            if column_index > 0:
                totals += [this_total]
            current_operation = operations[column_index]
            this_total = 0 if current_operation == '+' else 1
            operator = sum if current_operation == '+' else math.prod

        remade_number = ''.join(line[column_index] for line in number_lines)

        if remade_number.replace(' ', '') != '':
            this_total = operator([this_total, int(remade_number)])

    # The last total has not yet been added, so add it to the final result
    return totals + [this_total]
```

```python
example = '''123 328  51 64 
 45 64  387 23 
  6 98  215 314
*   +   *   +  
'''

new_maths('*   +   *   +  ', ['123 328  51 64 ', ' 45 64  387 23 ', '  6 98  215 314'])
```
