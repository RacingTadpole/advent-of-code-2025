```python
with open('input.txt', 'r') as f:
    raw_lines = f.readlines()

lines = [line.replace('\n', '') for line in raw_lines]

shapes = [lines[1:4], lines[6:9], lines[11:14], lines[16:19], lines[21:24], lines[26:29]]
print(shapes)

spaces = [{
    'width': int(line.split('x')[0]),
    'height': int(line.split('x')[1].split(':')[0]),
    'shapes': tuple(int(n) for n in line.split(' ')[1:])
    } for line in lines[30:]]
print(spaces[:3])
```

```python
shape_sizes = [sum(s == '#' for row in shape for s in row) for shape in shapes]
print(shape_sizes)
```

```python
left_over_space = [sp['width'] * sp['height'] - sum(shape_sizes[shape_index] * num
                    for shape_index, num in enumerate(sp['shapes']))
                    for sp in spaces]

sum([l >= 0 for l in left_over_space])
```
