```python
with open('input.txt', 'r') as f:
    raw_lines = f.readlines()

positions = [[int(c) for c in line.replace('\n', '').split(',')] for line in raw_lines]
```

```python
# Do the simplest thing and search for the closest two in a nested for loop.
def get_next_closest_pair(positions, pairs):
    closest_sqr_dist = 1e16
    closest_pair = []

    for (i, (x1, y1, z1)) in enumerate(positions):
        for (j_offset, (x2, y2, z2)) in enumerate(positions[i + 1:]):
            j = i + 1 + j_offset
            if (i, j) not in pairs:
                d = (x1 - x2) ** 2 + (y1 - y2) ** 2 + (z1 - z2) ** 2
                if d < closest_sqr_dist:
                    closest_sqr_dist = d
                    closest_pair = (i, j)

    return closest_pair
```

```python
def get_pairs(positions, num):
    pairs = []
    for i in range(num):
        if i % 10 == 0:
            print(i)
        closest_pair = get_next_closest_pair(positions, pairs)
        # print(positions[closest_pair[0]], positions[closest_pair[1]])
        pairs += [closest_pair]

    return pairs
```

```python
pairs = get_pairs(positions, 1000)
# pairs
```

```python
def get_nontrivial_circuits(pairs):
    circuits = []

    for pair in pairs:
        found_both = False
        found_one = None
        print(circuits)
        for i in range(len(circuits)):
            circuit = circuits[i]
            if pair[0] in circuit and pair[1] in circuit:
                found_both = True
                break
            if pair[0] in circuit or pair[1] in circuit:
                if found_one is None:
                    found_one = i
                    circuits[i] = circuit.union((pair[1 if pair[0] in circuit else 0],))
                else:
                    found_both = True
                    circuits[found_one] = circuits[found_one].union(circuit)
                    del circuits[i]
                    break

        if found_one is None and not found_both:
            circuits += [set(pair)]

    return sorted(circuits, key=lambda s: -len(s))
```

```python
final_circuits = get_nontrivial_circuits(pairs)
```
