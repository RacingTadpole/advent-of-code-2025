```python
import itertools
```

# Part 1

```python
with open('input.txt', 'r') as f:
    raw_lines = f.readlines()

targets = [tuple(1 if value=='#' else 0 for value in line.split(' ')[0][1:-1]) for line in raw_lines]
buttonses = [[tuple(int(b) for b in button[1:-1].split(',')) for button in line.split(' ')[1:-1]] for line in raw_lines]
```

```python
out_dict = {}
for N in range(1, 10):
    print(f'Checking {N} presses')
    for i in range(len(targets)):
        if i not in out_dict:
            target = targets[i]
            buttons = buttonses[i]
            states = [tuple(sum(1 for button in combo if slot in button) % 2 for slot in range(len(target)))
                for combo in itertools.combinations(buttons, N)]

            if any([state == target for state in states]):
                out_dict[i] = N

    if len(out_dict) == len(targets):
        break
```

```python
from scipy.optimize import linprog
```

```python
with open('input.txt', 'r') as f:
    raw_lines = f.readlines()

targets = [tuple(int(value) for value in line.split(' ')[-1][1:-2].split(',')) for line in raw_lines]
buttonses = [[tuple(int(b) for b in button[1:-1].split(',')) for button in line.split(' ')[1:-1]] for line in raw_lines]
```

```python
def transpose(M):
    return [[M[j][i] for j in range(len(M))] for i in range(len(M[0]))]
```

```python
def get_optimal(target, buttons):
    # The objective function we want to minimize is just the total number of presses
    c = [1] * len(buttons)

    # Equality constraint matrix
    At = [tuple(1 if n in b else 0 for n in range(len(target))) for b in buttons]
    A = transpose(At)

    # and vector
    b = target

    # solve 
    results = linprog(c=c, A_eq=A, b_eq=b, method='highs', integrality=True)

    # print results
    if results.status != 0:
        print('Could not find optimal solution!')

    return results
```

```python
out = []

for i in range(len(targets)):
    target = targets[i]
    buttons = buttonses[i]

    results = get_optimal(target, buttons)

    out+= [results.fun]
```
