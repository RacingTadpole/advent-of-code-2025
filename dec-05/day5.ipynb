# Part 1

```python
def get_fresh_ranges(input):
    return [{'from': int(r.split('-')[0]), 'to': int(r.split('-')[1])} for r in input if '-' in r]

def get_ingredients(input):
    return [int(r) for r in input if '-' not in r and len(r) > 0]
```

```python
get_fresh_ranges(example), get_ingredients(example)
```

```python
def count_fresh(data):
    fresh_count = 0
    for id in get_ingredients(data):
        for fresh_range in get_fresh_ranges(data):
            if id >= fresh_range['from'] and id <= fresh_range['to']:
                fresh_count += 1
                break

    return fresh_count
```

```python
count_fresh(example)
```

```python
count_fresh(all_lines)
```

# Part 2

```python
def join_ranges(fresh_ranges):
    out = fresh_ranges[:]
    
    i = 0
    while i < len(out):
        j = i + 1
        while j < len(out):
            r1 = out[i]
            r2 = out[j]
            if r2['from'] <= r1['to'] and r2['from'] >= r1['from']:
                out[i] = {'from': r1['from'], 'to': r2['to']}
                # print(f'joined rows {i} & {j} {r1['from']}-{r2['from']}-{r1['to']}-{r2['to']}')
                del out[j]
            else:
                j += 1
        
        i += 1

    return out

def repeatedly_join(fresh_ranges):
    old = None
    current = fresh_ranges

    while current != old:
        old = current
        current = join_ranges(current)
    
    return current
```

```python
repeatedly_join(get_fresh_ranges(example))
```

```python
joined_all = repeatedly_join(get_fresh_ranges(all_lines))
```

```python
def size_all_ranges(joined_ranges):
    return [r['to'] - r['from'] + 1 for r in joined_ranges]
```

```python
example_answer = size_all_ranges(repeatedly_join(get_fresh_ranges(example)))
example_answer, sum(example_answer)
```
