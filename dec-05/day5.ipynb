# Part 1

```python
def get_fresh_ranges(input):
    return [{'from': int(r.split('-')[0]), 'to': int(r.split('-')[1])} for r in input if '-' in r]

def get_ingredients(input):
    return [int(r) for r in input if '-' not in r and len(r) > 0]
```

```python
get_fresh_ranges(example), get_ingredients(example)
```

```python
def count_fresh(data):
    fresh_count = 0
    for id in get_ingredients(data):
        for fresh_range in get_fresh_ranges(data):
            if id >= fresh_range['from'] and id <= fresh_range['to']:
                fresh_count += 1
                break

    return fresh_count
```

```python
count_fresh(example)
```

```python
count_fresh(all_lines)
```

# Part 2

```python
def join_ranges(fresh_ranges):
    sorted_ranges = sorted(fresh_ranges, key=lambda r: r['from'])
    joined = [sorted_ranges[0]]

    for r in sorted_ranges[1:]:
        if r['from'] <= joined[-1]['to']:
            joined[-1]['to'] = max(joined[-1]['to'], r['to'])
        else:
            joined += [r]
            
    return joined
```

```python
joined_example = join_ranges(get_fresh_ranges(example))
```

```python
joined_all = join_ranges(get_fresh_ranges(all_lines))
```

```python
def size_all_ranges(joined_ranges):
    return [r['to'] - r['from'] + 1 for r in joined_ranges]
```

```python
example_answer = size_all_ranges(joined_example)
example_answer, sum(example_answer)
```

```python
answer = size_all_ranges(joined_all)
sum(answer)
```
