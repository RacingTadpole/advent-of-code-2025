```python
with open('input.txt', 'r') as f:
    raw_lines = f.readlines()

lines = [[c for c in line.replace('\n', '').split(' ')] for line in raw_lines]

lines[0]
```

```python
path = {}

for line in lines:
    path[line[0][:-1]] = line[1:]
```

# Part 1

```python
import functools
```

```python
@functools.lru_cache(maxsize=512)
def make_node(label, final='out'):
    return {
        'label': label,
        'children': None if label == final else [make_node(child_label) for child_label in path[label]]
    }
```

```python
tree = make_node('you')
```

```python
def count_leaves(node):
    if node['children'] is None:
        return 1
    
    return sum(count_leaves(child) for child in node['children'])
```

```python
count_leaves(tree)
```

# Part 2

```python
from dataclasses import dataclass, field

# We need to do this because we cannot hash a dictionary.
# But a node is uniquely defined by its label, so use that as the hash.
@dataclass
class Node():
    label: str
    children: list('Node') = field(default_factory=list)

    def __hash__(self):
        # Add one so the hash is different from the string itself, why not...
        return hash(self.label) + 1
```

```python
@functools.lru_cache(maxsize=512)
def make_node2(label, final='out'):
    return Node(label, [] if label == final else [make_node2(child_label) for child_label in path[label]])
```

```python
svr_tree = make_node2('svr')
```

```python
@functools.lru_cache(maxsize=2048)
def count_leaves2(node, required_labels=(), required_labels_so_far=()):
    if len(node.children) == 0 and all(r in required_labels_so_far for r in required_labels):
        return 1
    
    return sum(count_leaves2(child, required_labels, 
        tuple(required_labels_so_far + 
            ((node.label,) if node.label in required_labels and node.label not in required_labels_so_far else ()))
        )
        for child in node.children)
```

```python
count_leaves2(svr_tree, required_labels=('dac', 'fft'))
```
